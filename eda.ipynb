```python
import pandas as pd
import numpy as np

import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("datasets/train.csv")

columns = df.columns.to_list()
subjects = columns[6:]
df["mean"] = df[subjects].mean(axis=1)

df
```

```python
df.describe()
```

```python
# null cells
df.isna().sum()
```

```python
custom_palette = ["blue", "green", "red", "orange"]
x = "Hogwarts House"
plt.title("House Dist")
sns.countplot(df, x=x, hue=x, palette=custom_palette)
```

```python
plt.title("Hands")
sns.countplot(df, x="Best Hand")
```

```python
plt.title("Mean Dist")
sns.histplot(df, x="mean", hue="Hogwarts House", palette=custom_palette)
```

```python
for s in subjects:
    plt.title(s)
    sns.histplot(df, x=s, hue="Hogwarts House", palette=custom_palette)
    plt.show()
```
